```python
%%time
latS = -70
latN = 20
lonW = 90-360
lonE = 180-360
ekey='1deg'
expt = exptdata.exptdict[ekey]['expt']
print(expt)

sea_level_z0_1_mod_month = cc.querying.getvar(expt=expt,
                                     session=session,
                                     ncfile='ocean_month.nc',
                                     variable='eta_t').\
sel(yt_ocean=slice(latS,latN)).\
sel(xt_ocean=slice(lonW,lonE))
print(sea_level_z0_1_mod_month)

sea_level_z0_1_mod_month.load()
print(sea_level_z0_1_mod_month)

xt_ocean = sea_level_z0_1_mod_month.xt_ocean
xt_ocean_corrected = xt_ocean + 360
sea_level_z0_1_mod_month = sea_level_z0_1_mod_month.assign_coords(xt_ocean=xt_ocean_corrected)
print(sea_level_z0_1_mod_month)

sea_level_z0_1_mod = sea_level_z0_1_mod_month.groupby('time.year').mean('time')
print(sea_level_z0_1_mod)

sea_level_z0_1_mod = sea_level_z0_1_mod.sel(year=slice(1993+60*4,2017+60*4))
print(sea_level_z0_1_mod)
```

```python
year = sea_level_z0_1_mod.year
lat_1_mod = sea_level_z0_1_mod.yt_ocean
lon_1_mod = sea_level_z0_1_mod.xt_ocean

sea_level_z0_1_trans_mod = sea_level_z0_1_mod
print(np.shape(sea_level_z0_1_trans_mod))
sea_level_z0_1_slope_mod = xr.DataArray(
    np.zeros([np.shape(lat_1_mod)[0], np.shape(lon_1_mod)[0]]), dims=(
    'lat_1', 'lon_1'), coords=[lat_1_mod, lon_1_mod], name='sea_level_z0_1')
sea_level_z0_1_p_value_mod = xr.DataArray(
    np.zeros([np.shape(lat_1_mod)[0], np.shape(lon_1_mod)[0]]), dims=(
    'lat_1', 'lon_1'), coords=[lat_1_mod, lon_1_mod], name='sea_level_z0_1')
sea_level_z0_1_std_err_mod = xr.DataArray(
    np.zeros([np.shape(lat_1_mod)[0], np.shape(lon_1_mod)[0]]), dims=(
    'lat_1', 'lon_1'), coords=[lat_1_mod, lon_1_mod], name='sea_level_z0_1')
for iid, i in enumerate(lat_1_mod):
    for jid, j in enumerate(lon_1_mod):
        sea_level_z0_1_slope_mod[iid, jid], _, _, sea_level_z0_1_p_value_mod[iid, jid], sea_level_z0_1_std_err_mod[iid, jid] = \
        stats.linregress(year, sea_level_z0_1_trans_mod[:, iid, jid])
    print('lat_1_mod ' + str(np.array(i)))
print(sea_level_z0_1_slope_mod)
```
