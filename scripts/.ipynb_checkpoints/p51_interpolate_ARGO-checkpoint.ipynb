```python
input_path = '/g/data/e14/erd561/NOAA_OI_SST_V2/'

output_path = '/g/data/e14/erd561/Australia_3/'
```

```python
year = np.arange(1993,2018)

lonW = 88
lonE = 182

latS = -72
latN = 22

temp_z0 = xr.open_dataset(input_path + 'sst.mnmean.nc').\
                sst.\
                sel(lon=slice(lonW,lonE)).\
                sel(lat=slice(latS,latN,-1)).\
                groupby('time.year').\
                mean('time', skipna=True).\
                sel(year=slice(1993,2017))

temp_z0 = temp_z0.assign_coords(lon=temp_z0.lon-360)

mask = xr.open_dataset(input_path + 'lsmask.nc').\
                mask.\
                sel(lon=slice(lonW,lonE)).\
                sel(lat=slice(latS,latN,-1))

mask = mask.assign_coords(lon=mask.lon-360)

temp_z0 = temp_z0.where(mask==1, np.nan)

temp_z0 = temp_z0.squeeze()
print(temp_z0)
```

```python
lat = temp_z0.lat
lon = temp_z0.lon

# temp_z0_trans = np.transpose(temp_z0, (2,1,0))
temp_z0_trans = temp_z0
print(np.shape(temp_z0_trans))
temp_z0_slope = xr.DataArray(
    np.zeros([np.shape(lat)[0], np.shape(lon)[0]]), dims=(
    'lat', 'lon'), coords=[lat, lon], name='temp_z0')
temp_z0_p_value = xr.DataArray(
    np.zeros([np.shape(lat)[0], np.shape(lon)[0]]), dims=(
    'lat', 'lon'), coords=[lat, lon], name='temp_z0')
temp_z0_std_err = xr.DataArray(
    np.zeros([np.shape(lat)[0], np.shape(lon)[0]]), dims=(
    'lat', 'lon'), coords=[lat, lon], name='temp_z0')
for iid, i in enumerate(lat):
    for jid, j in enumerate(lon):
        temp_z0_slope[jid, iid], _, _, temp_z0_p_value[jid, iid], temp_z0_std_err[jid, iid] = \
        stats.linregress(year, temp_z0_trans[:, jid, iid])
    print('lat ' + str(np.array(i)))
print(temp_z0_slope)
```

```python
# literally copy-pasted from 
# /g/data/hh5/tmp/cosima/access-om2/1deg_jra55v13_iaf_spinup1_B1/output059/ocean

lonW = 90
lonE = 180

latS = -70
latN = 20

lat_1 = xr.open_dataset(
'/g/data/hh5/tmp/cosima/access-om2/1deg_jra55v13_iaf_spinup1_B1/output059/ocean/ocean.nc').\
yt_ocean.sel(yt_ocean=slice(latS,latN)).values

lon_1 = xr.open_dataset(
'/g/data/hh5/tmp/cosima/access-om2/1deg_jra55v13_iaf_spinup1_B1/output059/ocean/ocean.nc').\
xt_ocean.sel(xt_ocean=slice(lonW-360,lonE-360)).values

grid_lon_1, grid_lat_1 = np.meshgrid(lon_1,lat_1)

temp_z0_grid_lon, temp_z0_grid_lat = np.meshgrid(temp_z0.lon, temp_z0.lat)
temp_z0_lon_flat = np.transpose(temp_z0_grid_lon.flatten('F'))
temp_z0_lat_flat = temp_z0_grid_lat.flatten('F')
temp_z0_flat_list = [temp_z0_lon_flat, temp_z0_lat_flat]    
temp_z0_grid_points = np.transpose(temp_z0_flat_list)

for y in year:
    temp_z0_flatten = temp_z0.sel(year=y).stack(z=('lon', 'lat'))

    temp_z0_1_now = interpolate.griddata(temp_z0_grid_points, temp_z0_flatten, 
                                 (grid_lon_1, grid_lat_1), 
                                 method='linear')
    
    if y == 1993:
        temp_z0_1 = temp_z0_1_now
    else:
        temp_z0_1 = np.dstack((temp_z0_1, temp_z0_1_now))
        
    print(y)

print(np.shape(temp_z0_1))
```

```python
temp_z0_1_trans = np.transpose(temp_z0_1, (2,1,0))
print(np.shape(temp_z0_1_trans))
temp_z0_1_slope = xr.DataArray(
    np.zeros([np.shape(lat_1)[0], np.shape(lon_1)[0]]), dims=(
    'lat_1', 'lon_1'), coords=[lat_1, lon_1], name='temp_z0_1')
temp_z0_1_p_value = xr.DataArray(
    np.zeros([np.shape(lat_1)[0], np.shape(lon_1)[0]]), dims=(
    'lat_1', 'lon_1'), coords=[lat_1, lon_1], name='temp_z0')
temp_z0_1_std_err = xr.DataArray(
    np.zeros([np.shape(lat_1)[0], np.shape(lon_1)[0]]), dims=(
    'lat_1', 'lon_1'), coords=[lat_1, lon_1], name='temp_z0')
for iid, i in enumerate(lat_1):
    for jid, j in enumerate(lon_1):
        temp_z0_1_slope[iid, jid], _, _, temp_z0_1_p_value[iid, jid], temp_z0_1_std_err[iid, jid] = \
        stats.linregress(year, temp_z0_1_trans[:, jid, iid])
    print('lat_1 ' + str(np.array(i)))
print(temp_z0_1_slope)
```
